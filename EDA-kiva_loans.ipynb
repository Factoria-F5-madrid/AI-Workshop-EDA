# Tomamos las columnas de fecha y les definimos el tipo de dato "**datetime**"

```python
# Convertir a datetime las fechas
kiva_loans_df["posted_time"] = pd.to_datetime(kiva_loans_df["posted_time"], errors="coerce")
kiva_loans_df["disbursed_time"] = pd.to_datetime(kiva_loans_df["disbursed_time"], errors="coerce")

# Normalizar a la fecha (eliminar la la hora, establecindola en medianoche) manteniendo el tipo datetime
kiva_loans_df["posted_time"] = kiva_loans_df["posted_time"].dt.normalize()
kiva_loans_df["disbursed_time"] = kiva_loans_df["disbursed_time"].dt.normalize()
```

# Nos aseguramos que los datos de tipo "str" no tengan espacios innecesarios y que estn escritas en minscula

```python
def clean_text(columna_texto):
    return columna_texto.astype(str).str.strip().str.lower()

for col in kiva_loans_df.columns:
    if kiva_loans_df[col].dtype == 'string':
        kiva_loans_df[col] = clean_text(kiva_loans_df[col])
```

# Clasificacin de prstamos segn la fecha de desembolso
 Creamos la columna `loan_type` para saber si un prstamo fue desembolsado
 antes o despus de la fecha de publicacin:
- "pre_disbursed": el prstamo se desembols **antes** de publicarse (disbursed_time < posted_time)
- "post_disbursed": el prstamo se desembols **despus** de publicarse (disbursed_time >= posted_time)

```python
kiva_loans_df["loan_type"] = np.where(
    kiva_loans_df["disbursed_time"] < kiva_loans_df["posted_time"],
    "pre_disbursed",
    "post_disbursed"
)
```

# Verificamos la existencia de outliers (valores atpicos), respecto a las columnas **funded_amount:** y **loan_amount**

 **Regla de Tukey:** La Regla de Tukey es una forma sencilla de detectar valores atpicos (outliers) usando cuartiles.
La idea es ver qu tan lejos estn los datos de la parte **normal** del conjunto.

- Q1 = valor donde termina el 25% ms bajo de los datos

- Q3 = valor donde termina el 75% de los datos

- IQR = Q3  Q1  mide la zona donde vive la mitad central de los datos

Un dato se considera outlier si est demasiado lejos de esa zona:

- Lmite inferior = Q1  1.5  IQR

- Lmite superior = Q3 + 1.5  IQR

Todo lo que quede fuera de ese rango se marca como valor atpico.

** Quin fue Tukey?

John Tukey fue un estadstico estadounidense, pionero en anlisis y exploracin de datos. Es conocido por inventar el boxplot y acuar el trmino bit en informtica.

```python
kiva_loans_df['loan_amount_category'].value_counts()
```

```python
categories = ["micro", "small", "medium", "large"]

for cat in categories:
    df_cat = kiva_loans_df[kiva_loans_df["loan_amount_category"] == cat]

    if df_cat.empty:
        print(f"\n{cat}: (sin datos)")
        continue

    q1 = df_cat["loan_amount"].quantile(0.25)
    q3 = df_cat["loan_amount"].quantile(0.75)
    iqr = q3 - q1

    # Limites para detectar outliers
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    # Filtrar outliers
    outliers = df_cat[(df_cat["loan_amount"] < lower_bound) | (df_cat["loan_amount"] > upper_bound)]
    n_outliers = len(outliers)

    print(f"\nCategora: {cat}")
    print(f"  Q1: {q1}")
    print(f"  Q3: {q3}")
    print(f"  IQR: {iqr}")
    print(f"  Outliers: {n_outliers}")
```
